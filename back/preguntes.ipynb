```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os

# Función para leer el archivo JSON
def read_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    return data

# Función para calcular el porcentaje de respuestas correctas por pregunta
def calcular_porcentajes(data):
    porcentajes_correctos = []
    preguntas = []
    
    for pregunta in data['preguntes']:
        total_respuestas = sum([r['contador'] for r in pregunta['resposta_usuaris']])
        correctas = sum([r['contador'] for r in pregunta['resposta_usuaris'] if r['resposta'] == pregunta['resposta_correcta']])
        
        if total_respuestas > 0:
            porcentaje_correcto = (correctas / total_respuestas) * 100
        else:
            porcentaje_correcto = 0
        
        preguntas.append(pregunta['pregunta'])
        porcentajes_correctos.append(porcentaje_correcto)
    
    return preguntas, porcentajes_correctos

# Función para generar la gráfica
def generar_grafica(preguntas, porcentajes):
    x = np.arange(len(preguntas))
    fig, ax = plt.subplots(figsize=(12, 8))

    # Crear barras
    ax.barh(x, porcentajes, color='green', alpha=0.6)
    
    # Añadir etiquetas de datos y personalización del gráfico
    for i, v in enumerate(porcentajes):
        ax.text(v + 1, i, f'{v:.2f}%', color='black', va='center')
    
    ax.set_yticks(x)
    ax.set_yticklabels(preguntas, fontsize=10, ha='right')
    ax.invert_yaxis()  # Invertir el eje Y para que la primera pregunta esté en la parte superior

    # Etiquetas y título
    ax.set_xlabel('Porcentaje de respuestas correctas', fontsize=12)
    ax.set_title('Porcentaje de Aciertos por Pregunta', fontsize=14)

    plt.tight_layout()
    plt.show()

# Función principal
def main():
    # Ruta al archivo JSON (modificar según tu ruta local)
    file_path = os.path.join('ruta', 'a', 'tu', 'archivo', 'Respostes_usuarisBase.json')
    
    # Leer y procesar los datos
    data = read_json(file_path)
    preguntas, porcentajes_correctos = calcular_porcentajes(data)
    
    # Generar la gráfica
    generar_grafica(preguntas, porcentajes_correctos)

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
