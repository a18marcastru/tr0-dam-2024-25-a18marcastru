```python
import json
import matplotlib.pyplot as plt
import os

# Función para leer el archivo JSON
def read_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    return data

# Función para encontrar la pregunta solicitada
def buscar_pregunta(data, pregunta_texto):
    for pregunta in data['preguntes']:
        if pregunta['pregunta'].lower() == pregunta_texto.lower():
            return pregunta
    return None

# Función para generar el gráfico de torta
def generar_grafico_torta(pregunta_data):
    respuestas = [r['resposta'] for r in pregunta_data['resposta_usuaris']]
    contadores = [r['contador'] for r in pregunta_data['resposta_usuaris']]
    correcta = pregunta_data['resposta_correcta']

    # Colores para las respuestas, destacando la correcta
    colors = ['lightcoral' if r != correcta else 'limegreen' for r in respuestas]

    # Crear el gráfico de torta
    fig, ax = plt.subplots(figsize=(8, 8))
    wedges, texts, autotexts = ax.pie(
        contadores, labels=respuestas, autopct='%1.1f%%', colors=colors,
        startangle=90, wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}
    )

    # Destacar la respuesta correcta (desplazamiento)
    correcta_idx = respuestas.index(correcta)
    wedges[correcta_idx].set_explode([0.1])  # Separar la parte correspondiente a la respuesta correcta

    # Personalizar los textos de la gráfica
    for text in autotexts:
        text.set_color('black')
        text.set_fontsize(10)

    # Título y leyenda
    ax.set_title(f"Respostes per a la pregunta: \n{pregunta_data['pregunta']}", fontsize=14)
    plt.show()

# Función principal
def main():
    # Ruta al archivo JSON (modificar según tu ruta local)
    file_path = os.path.join('ruta', 'a', 'tu', 'archivo', 'Respostes_usuarisBase.json')
    
    # Leer el archivo JSON
    data = read_json(file_path)

    # Pedir al usuario la pregunta que desea consultar
    pregunta_texto = input("Escriu la pregunta de la qual vols veure les dades: ")
    
    # Buscar la pregunta en el JSON
    pregunta_data = buscar_pregunta(data, pregunta_texto)
    
    if pregunta_data:
        # Generar el gráfico de torta si se encuentra la pregunta
        generar_grafico_torta(pregunta_data)
    else:
        print(f"No s'ha trobat la pregunta: {pregunta_texto}")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
